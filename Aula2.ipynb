```python
#puxar media
media_buscagoogle = df.groupby('BANCO')['BUSCAS NO GOOGLE'].mean()
media_buscagoogle
```

```python
plt.figure(figsize=(10,5))
media_buscagoogle.plot(kind='bar', color='lightgreen')
plt.xlabel('BANCOS')
plt.ylabel('BUSCAS NO GOOGLE')
plt.title('Media de buscas no google por banco')
plt.show()
```

```python
#puxar variancia (a variancia indica o quanto os dados vao ficando longe da media)
variancia_buscagoogle = df.groupby('BANCO')['BUSCAS NO GOOGLE'].var()
variancia_buscagoogle
```

```python
plt.figure(figsize=(10,5))
variancia_buscagoogle.plot(kind='bar', color='pink')
plt.xlabel('BANCOS')
plt.ylabel('BUSCAS NO GOOGLE')
plt.title('Variancia de buscas no google por banco')
plt.show()
```

```python
# Qual banco gerou uma maior média de depósitos? R: BANCO E
media_depositos = df.groupby('BANCO')['DEPOSITO (MM)'].mean()
media_depositos

plt.figure(figsize=(10,5))
media_depositos.plot(kind='bar', color='black')
plt.xlabel('BANCOS')
plt.ylabel('DEPOSITO')
plt.title('Media de depositos por banco')
plt.show()
```

```python
media_depositos
```

```python
# Qual banco obteve uma maior média de investimento em campanhas? R: BANCO B 34529.133333
media_invest= df.groupby('BANCO')['INVESTIMENTO TOTAL'].mean()
media_invest

plt.figure(figsize=(10,5))
media_invest.plot(kind='bar', color='royalblue')
plt.xlabel('BANCOS')
plt.ylabel('INVESTIMENTOS')
plt.title('Media de investimentos por banco')
plt.show()
```

```python
media_invest
```

```python
# Os dados para cada banco em investimento em campanhas estão homogêneos ou heterogêneos? B maior pico, ou seja os dados estao mto heterogeneos, tem variance grande. Banco A e C sao homogeneos, pouca variancia. O banco E nem investe por isso nao aparece.
variancia_invest= df.groupby('BANCO')['INVESTIMENTO TOTAL'].var()
variancia_invest

plt.figure(figsize=(10,5))
variancia_invest.plot(kind='bar', color='olive')
plt.xlabel('BANCOS')
plt.ylabel('INVESTIMENTOS')
plt.title('Media de investimentos por banco')
plt.show()
```
