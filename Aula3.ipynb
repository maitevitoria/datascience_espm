```python
df= pd.read_excel('Aula2_DadosModelo.xlsx')
df
```

```python
#y=f(x)

#criar variavel para armanezar o Y (variavel dependente) o que quero prever -> DEPOSITO
y=df['DEPOSITO (MM)']

#criar variavel para conter todas as variaveis independentes o X
X=df[['BANCO',	'BUSCAS NO GOOGLE',	'PALAVRA CHAVE ORGANICO',	'PALAVRA CHAVE PAGO',	'TRAFEGO ORGANICO',	'TRAFEGO PAGO',	'TRAFEGO TOTAL',	'CUSTO TRAFEGO PAGO',	'TV MERCHANDISING',	'TV ASSINATURA',	'TV ABERTA',	'REVISTA',	'OOH',	'RADIO',	'JORNAL',	'CINEMA',	'INVESTIMENTO TOTAL','DEPOSITO (MM)']]
```

```python
X= pd.get_dummies(data=X)
```

```python
#criar datasets de treino e teste - 30% está sendo separado para teste
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
modelo_regressao = LinearRegression()
#estimar o modelo linear
modelo_regressao.fit(X_train, y_train)
```

```python
print('R-quadrado:', modelo_regressao.score(X_train, y_train))
```

```python
predictions=modelo_regressao.predict(X_test)
predictions
```
